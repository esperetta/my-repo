Parse the Xref.text file to make a record for each line in separate lists

```python
import sys
import re
feature_AC = [] # create a list which we will use to store data
feature_AC_specialDR1 = []
feature_AC_specialDR2 = []
feature_AC_specialDR3 = []
feature_DR1 = []
feature_DR2 = []
feature_DR3 = [] 
numentries = 0
numentriesSpecialDR1 = 0
numentriesSpecialDR2 = 0
numentriesSpecialDR3 = 0
with open('Xref_September.txt', 'r') as infile:
    for line in infile:
        if line.startswith('AC'):
            AC = infile.readline().rstrip()         
            FROM = infile.readline().rstrip()
            DR1 = infile.readline().rstrip()        
            TO = infile.readline().rstrip()
            if 'TO' in TO:         
                #print(TO)
                DR2 = infile.readline().rstrip()
                if DR2.startswith("DR"):
                    fine =infile.readline().rstrip()
                    if fine.startswith('DR'):
                        feature_AC_specialDR2.append(AC)
                        numentriesSpecialDR2 = numentriesSpecialDR2 + 1
                        #print("found")
                    else:
                        feature_AC.append(AC)
                        feature_DR1.append(DR1)
                        feature_DR2.append(DR2) 
                        numentries = numentries + 1
                else:
                    feature_AC_specialDR3.append(AC)
                    numentriesSpecialDR3 = numentriesSpecialDR3 + 1
            else:
                feature_AC_specialDR1.append(AC)
                numentriesSpecialDR1 = numentriesSpecialDR1 + 1

print(feature_AC, feature_DR1, feature_DR2)
print(numentries, numentriesSpecialDR1, numentriesSpecialDR2, numentriesSpecialDR3)
print(feature_AC_specialDR1, feature_AC_specialDR2, feature_AC_specialDR3)
    
```

Exctract information new gene name from the feature_DR2 list.

```python
import re

myl = feature_DR2
newgene_name = []
for element in myl:
    m = re.match(r'DR.+[\w.]+;\s[\w.]+;\s+(\w.*)\.', element)
    if m:
        found = m.group(1)
        newgene_name.append(found)
#newgene_name
len(newgene_name)
```

```python
import re

myl = feature_DR2
print(myl)
newFGgene_name = []
for element in myl:
    m = re.match(r'DR.+[\w.]+;\s(FBgn\d+);\s+(\w.*)\.', element)
    if m:
        found = m.group(1)
        newFGgene_name.append(found)
#newFGgene_name
len(newFGgene_name)
```
